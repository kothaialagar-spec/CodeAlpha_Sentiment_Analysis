```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
reviews = {
    'Review': [
        'This product is amazing',
        'I love this item',
        'Very bad quality',
        'Worst purchase ever',
        'Average product',
        'Excellent performance',
        'Not worth the money',
        'Highly recommended',
        'Poor packaging',
        'Good value for money',
        'Fantastic experience',
        'Terrible customer support',
        'Very satisfied with the product',
        'Waste of money',
        'The quality is excellent',
        'Could be better',
        'Easy to use and reliable',
        'Disappointed with the service',
        'Amazing features and performance',
        'The product is okay'
    ]
}

df = pd.DataFrame(reviews)

print("Dataset Loaded Successfully")
print(df.head())
```

```python
def get_sentiment(text):

    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Review"].apply(get_sentiment)

print(df)
```

```python
sentiment_counts = df["Sentiment"].value_counts()

print("\nSentiment Summary")
print(sentiment_counts)
```

```python
plt.figure(figsize=(8,5))

sentiment_counts.plot(kind="bar")

plt.title("Customer Review Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")

plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%'
)

plt.title("Sentiment Distribution")

plt.show()
```

```python
positive = sentiment_counts.get("Positive", 0)
negative = sentiment_counts.get("Negative", 0)
neutral = sentiment_counts.get("Neutral", 0)

print("----- BUSINESS INSIGHTS -----")

print(f"Positive Reviews : {positive}")
print(f"Negative Reviews : {negative}")
print(f"Neutral Reviews  : {neutral}")

if positive > negative:
    print("\nOverall customer opinion is positive.")
else:
    print("\nCustomer satisfaction needs improvement.")

print("\nNegative reviews indicate areas where product quality and customer service can be improved.")
```
